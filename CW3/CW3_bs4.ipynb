```python
#引用爬取数据的包
import requests
from bs4 import BeautifulSoup
import re
```

```python
#设置访问网站的头部信息
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11',
'Accept':'text/html;q=0.9,*/*;q=0.8',
'Accept-Charset':'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
'Accept-Encoding':'gzip',
'Connection':'close',
'Referer':'http://www.baidu.com/link?url=_andhfsjjjKRgEWkj7i9cFmYYGsisrnm2A-TN3XZDQXxvGsM9k9ZZSnikW2Yds4s&amp;wd=&amp;eqid=c3435a7d00006bd600000003582bfd1f'
}
```

```python
house_list = []
for i in range(1, 11):
    url = 'https://bj.lianjia.com/ershoufang/pg' + str(i) + '/'  #列表页URL
    response = requests.get(url, headers=headers).text  #获取当前页内容  
    soup = BeautifulSoup(response, 'html.parser')  #解析页面内容
    houses = soup.find_all('li', class_="clear LOGCLICKDATA")
    for house in houses:
        houseInfo = []
        houseInfo.append(house.select('div[class="title"]')[0].get_text())
        houseInfo.append(re.findall('(\d+\.*?\d+)平米', house.select('div[class="houseInfo"]')[0].get_text())[0])
        house_list.append(houseInfo)
```

```python
#创建数据表（构造DataFrame&存csv）
import pandas as pd
sorted_house_list = sorted(house_list, key=lambda x: float(x[1]))
df = pd.DataFrame(sorted_house_list, columns=['简介', '面积'])
df.to_csv("bs4.csv",encoding='utf-8',index=False)
df
```
